```python
import numpy as np
import matplotlib.pyplot as plt
#Distribucion Exponencial
# Definir parámetros
tasa_llegada = 1 / 15  # Promedio de 1 llegada cada 15 minutos
horas_operacion = 12
minutos_operacion = horas_operacion * 60  # Convertir a minutos

# Simular los tiempos entre llegadas
tiempo_entre_llegadas = np.random.exponential(1 / tasa_llegada, minutos_operacion)

# Calcular los tiempos de llegada acumulados
tiempo_llegadas = np.cumsum(tiempo_entre_llegadas)

# Filtrar solo llegadas dentro del tiempo de operación
tiempo_llegadas = tiempo_llegadas[tiempo_llegadas <= minutos_operacion]

# Mostrar resultados
print("Tiempos entre llegadas de vehículos (minutos):", tiempo_entre_llegadas[:10])  # Mostrar los primeros 10
print("Total de vehículos llegados en 12 horas:", len(tiempo_llegadas))

# Graficar histograma de tiempos entre llegadas
plt.figure(figsize=(8, 5))
plt.hist(tiempo_entre_llegadas, bins=30, color='lightcoral', edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Tiempo entre llegadas (minutos)")
plt.ylabel("Frecuencia")
plt.title("Distribución de Tiempos entre Llegadas de Vehículos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
