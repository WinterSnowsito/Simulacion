```python
import numpy as np
import matplotlib.pyplot as plt
#Distribucion LogNormal
# Definir parámetros
media_espera = 10  # Tiempo promedio de espera en minutos
desviacion_espera = 2  # Desviación estándar en minutos
n_clientes = 50  # Número de clientes

# Convertir la media y desviación estándar a los parámetros de la distribución lognormal
mu = np.log(media_espera**2 / np.sqrt(desviacion_espera**2 + media_espera**2))
sigma = np.sqrt(np.log(1 + (desviacion_espera**2 / media_espera**2)))

# Generar tiempos de espera con distribución lognormal
tiempos_espera = np.random.lognormal(mu, sigma, n_clientes)

# Mostrar resultados
print("Tiempos de espera simulados (minutos):", tiempos_espera[:10])  # Mostrar los primeros 10

# Graficar histograma de tiempos de espera
plt.figure(figsize=(8, 5))
plt.hist(tiempos_espera, bins=20, color='seagreen', edgecolor='black', alpha=0.7, density=True)
plt.xlabel("Tiempo de espera (minutos)")
plt.ylabel("Densidad")
plt.title("Distribución LogNormal de Tiempos de Espera")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
