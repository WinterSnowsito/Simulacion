```python
import numpy as np
import matplotlib.pyplot as plt
#Distribucion binominal
# Definir par√°metros
n_pedidos = 20  # N√∫mero de pedidos por d√≠a
p_exito = 0.7   # Probabilidad de satisfacer un pedido
n_dias = 7      # Simulaci√≥n para una semana

# Simular la cantidad de pedidos satisfechos por d√≠a
datos_simulados = np.random.binomial(n_pedidos, p_exito, n_dias)

# Mostrar resultados
print("Pedidos satisfechos por d√≠a durante una semana:", datos_simulados)

# Graficar los resultados
plt.figure(figsize=(8, 5))
plt.bar(range(1, n_dias + 1), datos_simulados, color='skyblue', edgecolor='black')
plt.xlabel("D√≠a")
plt.ylabel("Pedidos Satisfechos")
plt.title("Simulaci√≥n de Pedidos Satisfechos por D√≠a")
plt.xticks(range(1, n_dias + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
#
"""
El uso de la distribuci√≥n binomial en esta simulaci√≥n permite modelar el proceso de satisfacci√≥n de pedidos en funci√≥n de una probabilidad de √©xito constante.

Este tipo de an√°lisis es crucial para la gesti√≥n de inventarios y la planificaci√≥n operativa, ya que ayuda a predecir la cantidad de pedidos que podr√°n ser satisfechos y permite anticiparse a posibles faltantes de stock. Con esta informaci√≥n, las empresas pueden optimizar sus niveles de inventario y mejorar la satisfacci√≥n del cliente reduciendo los tiempos de espera y las p√©rdidas por desabastecimiento.
"""

```
