```python
import numpy as np
import matplotlib.pyplot as plt
#Distribucion binominal
# Definir parámetros
n_pedidos = 20  # Número de pedidos por día
p_exito = 0.7   # Probabilidad de satisfacer un pedido
n_dias = 7      # Simulación para una semana

# Simular la cantidad de pedidos satisfechos por día
datos_simulados = np.random.binomial(n_pedidos, p_exito, n_dias)

# Mostrar resultados
print("Pedidos satisfechos por día durante una semana:", datos_simulados)

# Graficar los resultados
plt.figure(figsize=(8, 5))
plt.bar(range(1, n_dias + 1), datos_simulados, color='skyblue', edgecolor='black')
plt.xlabel("Día")
plt.ylabel("Pedidos Satisfechos")
plt.title("Simulación de Pedidos Satisfechos por Día")
plt.xticks(range(1, n_dias + 1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
#
"""
El uso de la distribución binomial en esta simulación permite modelar el proceso de satisfacción de pedidos en función de una probabilidad de éxito constante.

Este tipo de análisis es crucial para la gestión de inventarios y la planificación operativa, ya que ayuda a predecir la cantidad de pedidos que podrán ser satisfechos y permite anticiparse a posibles faltantes de stock. Con esta información, las empresas pueden optimizar sus niveles de inventario y mejorar la satisfacción del cliente reduciendo los tiempos de espera y las pérdidas por desabastecimiento.
"""

```
