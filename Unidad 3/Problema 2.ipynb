```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
lambda_rate = 1 / 15  # Tasa de llegada (1 llegada cada 15 minutos)
total_time = 12 * 60  # 12 horas en minutos

# Número de eventos (llegadas) a simular
n_events = int(total_time / 15)  # Aproximadamente el número de llegadas

# Generación de los tiempos entre llegadas usando la distribución exponencial
# Usamos np.random.exponential para generar los tiempos entre llegadas
inter_arrival_times = np.random.exponential(1 / lambda_rate, n_events)

# Calcular los tiempos de llegada acumulados
arrival_times = np.cumsum(inter_arrival_times)

# Filtrar los tiempos de llegada que ocurren dentro de las 12 horas de operación
arrival_times = arrival_times[arrival_times <= total_time]

# Mostrar los resultados
print(f"Tiempos de llegada de los vehículos (en minutos): {arrival_times}")
print(f"Número total de llegadas: {len(arrival_times)}")

# Graficar los tiempos de llegada
plt.figure(figsize=(10, 6))
plt.step(arrival_times, np.arange(1, len(arrival_times) + 1), where='post')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Número de llegada')
plt.title('Tiempos de llegada de vehículos eléctricos')
plt.grid(True)
plt.show()

# Conclusión:
"""
Este programa simula los tiempos entre llegadas de vehículos eléctricos a una estación de carga rápida durante un día de operación de 12 horas.
 La distribución exponencial se utiliza para modelar la variabilidad en los tiempos entre llegadas, lo que permite prever las necesidades de carga
 según el comportamiento de los vehículos. Este tipo de simulaciones puede ser útil para la planificación de la infraestructura de carga,
ayudando a estimar el número adecuado de puntos de carga necesarios para optimizar el servicio y reducir los tiempos de espera.
"""
```
