```python
import numpy as np
import matplotlib.pyplot as plt
mu = np.log(10)
sigma = 2 / 10
n_customers = 50

wait_times = np.random.lognormal(mu, sigma, n_customers)

# Mostrar los resultados
print(f"Tiempos de espera de los clientes (en minutos): {wait_times}")
print(f"Tiempo total de espera acumulado: {np.sum(wait_times):.2f} minutos")


plt.figure(figsize=(10, 6))
plt.hist(wait_times, bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Tiempo de espera (minutos)')
plt.ylabel('Número de clientes')
plt.title('Distribución de los tiempos de espera en el mostrador de servicio técnico')
plt.grid(True)
plt.show()

# Conclusión:
"""
Este programa simula los tiempos de espera de 50 clientes en el mostrador de servicio técnico de una tienda departamental de electrónica.
La distribución lognormal se utiliza para modelar los tiempos de espera, dada su característica de que el tiempo promedio de espera es de
10 minutos con una desviación estándar de 2 minutos. La simulación permite analizar la duración de la fila de espera, lo que es útil para
la gestión del personal y la planificación de tiempos de atención, para evitar largas esperas y mejorar la experiencia del cliente.
"""
```
